```python

def longestCommonPrefix(strs):
    prefix = ''
    len_prefix = 0
    length, min_str_len = len(strs), len(strs[0])

    if length == 1:
        return strs[0]
    
    while(len_prefix < min_str_len):
        for i in range(length-1):
            if strs[i][len_prefix] != strs[i+1][len_prefix]:
                return prefix
            
            if len_prefix == 0:
                next_str_len = len(strs[i+1])
                if min_str_len > next_str_len:
                    min_str_len = next_str_len

        prefix += strs[i][len_prefix]
        len_prefix += 1

    print(min_str_len)
    return prefix

strs = ['a']
longestCommonPrefix(strs)
```

```python

dict = {')':'(', '}':'{', ']':'['}

def isValid(s):
    if len(s)%2!=0:
            return False

    heap = []
    for x in s:
        if x == '(' or x == '[' or x == '{':
            heap += x
        elif len(heap)==0 or heap[-1]!=dict[x]:
            return False
        else:
            heap.pop()
    
    if len(heap)!=0:
        return False

    return True
```
